```python
import pandas as pd
```

```python
df = pd.read_csv('SortingSignalsSwissprot.csv')
```

```python
df.rename(columns={'Sequence':'Sequence_full'}, inplace=True)
```

```python
df
```

```python
new_df = df[df['Types']=='MT']
```

```python
df['Sequence_full'] = df['Sequence_full'].apply(list)
```

```python

# Define a function to extract begin and end positions of the motif
def extract_motif_positions(sequence, motif_positions):
    # Convert positions to integers
    motif_positions = [int(pos) for pos in motif_positions]
    
    # Construct the motif from the sequence and motif_positions
    motif = ''.join(sequence[pos-1] if pos != 0 else '-' for pos in motif_positions)
    
    # Initialize start and end positions
    start_pos = None
    end_pos = None
    
    # If motif is not just a dash
    if motif != '-':
        # Iterate through the sequence to find the motif
        for i in range(len(sequence) - len(motif) + 1):
            if sequence[i:i+len(motif)] == motif:
                start_pos = i + 1
                end_pos = i + len(motif)
                break
    
    return start_pos, end_pos
```

```python
# Apply the function to create new columns with extracted information
new_df['Begin'], new_df['End'] = zip(*new_df.apply(lambda row: extract_motif_positions(row['Sequence_full'], row['AnnotEncoded']), axis=1))
```
